```python
#Bubble Sort:

def bubbleSort(lista):
  numeros = len(lista)
  for numero in range(numeros):
    for index in range(0, numeros-numero-1):
      if lista[index] > lista [index + 1]:
        lista[index], lista[index+1] = lista[index+1], lista[index]
        print(f"Troca: {lista}")
    print(f"\nPassada {numero + 1}: {lista}")
  return lista

listaExemplo = [64, 34, 25, 12, 22, 11, 90]
print("Lista Original: ")
print(listaExemplo)
print("\nOrdenando...\n")
listaOrdenada = bubbleSort(listaExemplo)
print("\nLista Ordenada:")
print(listaOrdenada)
```

```python
#Insert Sort
def insertionSort(numeros):
  for numero in range(1, len(numeros)):
    chave = numeros[numero]
    numAnterior = numero - 1
    while numAnterior >= 0 and numeros[numAnterior] > chave:
      numeros[numAnterior + 1] = numeros[numAnterior]
      numAnterior -= 1
      print(f"movimentação: {numeros}")
    numeros[numAnterior + 1] = chave
    print(f"\ninserção: {numero}: {numeros}")
  return numeros

listaExemplo = [64, 34, 25, 12, 22, 11, 90]
print("Lista Original: ")
print(listaExemplo)
print("\nOrdenando...\n")
listaOrdenada = insertionSort(listaExemplo)
print("\nLista Ordenada:")
print(listaOrdenada)
```

```python
#Select Sort

def selectionSort(numeros):
  for numero in range(len(numeros)):
    min = numero
    for num in range(numero + 1, len(numeros)):
      if numeros[num] < numeros[min]:
        min = num
    numeros[numero], numeros[min] = numeros[min], numeros[numero]
    print(f"Troca: {numero + 1}: {numeros}")
  return numeros

listaExemplo = [64, 34, 25, 12, 22, 11, 90]
print("Lista Original: ")
print(listaExemplo)
print("\nOrdenando...\n")
listaOrdenada = selectionSort(listaExemplo)
print("\nLista Ordenada:")
print(listaOrdenada)
```

```python
#Merge Sort

def mergeSort(numeros):
  if len(numeros) > 1:
    meio = len(numeros) // 2
    esquerda = numeros[:meio]
    direita = numeros[meio:]

    mergeSort(esquerda)
    mergeSort(direita)


    i = j = k = 0

    while i < len(esquerda) and j < len(direita):
      if esquerda[i] < direita[j]:
        numeros[k] = esquerda[i]
        i += 1
      else:
        numeros[k] = direita[j]
        j += 1
      k += 1

    while i < len(esquerda):
      numeros[k] = esquerda[i]
      i += 1
      k += 1

    while j < len(direita):
      numeros[k] = direita[j]
      j += 1
      k += 1

    print(f"Merge: {numeros}")
  return numeros

listaExemplo = [64, 34, 25, 12, 22, 11, 90]
print("Lista Original: ")
print(listaExemplo)
print("\nOrdenando...\n")
listaOrdenada = mergeSort(listaExemplo)
print("\nLista Ordenada:")
print(listaOrdenada)
```

```python
#Quick Sort

def quickSort(numeros):
  def particiona(numeros, baixo, alto):
    pivo = numeros[alto]
    numero = baixo - 1

    for num in range(baixo,alto):
      if numeros[num] <= pivo:
        numero += 1
        numeros[numero], numeros[num] = numeros[num], numeros[numero]
    numeros[numero + 1], numeros[alto] = numeros[alto], numeros[numero + 1]
    return numero + 1

  def quickSortRecursivo(numeros, baixo, alto):
    if baixo < alto:
      pi = particiona(numeros, baixo, alto)
      quickSortRecursivo(numeros, baixo, pi - 1)
      quickSortRecursivo(numeros, pi + 1, alto)

  quickSortRecursivo(numeros, 0, len(numeros) - 1)
  return numeros

listaExemplo = [64, 34, 25, 12, 22, 11, 90]
print("Lista Original: ")
print(listaExemplo)
print("\nOrdenando...\n")
listaOrdenada = quickSort(listaExemplo)
print("\nLista Ordenada:")
print(listaOrdenada)
```

```python
#Shell sort

def shellSort(lista):
  n = len(lista)
  intervalo = n // 2

  while intervalo > 0:
    for i in range(intervalo, n):
      temp = lista[i]
      j = i
      while j >= intervalo and lista[j - intervalo] > temp:
        lista[j] = lista[j - intervalo]
        j -= intervalo
      lista[j] = temp
    print(f"\nIntervalo {intervalo} : {lista}")
    intervalo //= 2
  return lista

listaExemplo = [64, 34, 25, 12, 22, 11, 90]
print("Lista Original: ")
print(listaExemplo)
print("\nOrdenando...\n")
listaOrdenada = shellSort(listaExemplo)
print("\nLista Ordenada:")
print(listaOrdenada)
```
